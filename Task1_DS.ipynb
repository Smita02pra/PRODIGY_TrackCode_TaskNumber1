```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv("world population.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df=df.drop(columns=['CCA3','Rank'])
df
```

```python
df=df.sort_values('2022 Population',ascending=False)
df
```

```python
plt.figure(figsize=(10, 5))

plt.bar(df['Country/Territory'][-10:], df['1970 Population'][-10:])
plt.title('Least Populated Countries [2022]')
plt.xlabel('Countries')
plt.ylabel('Population')
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.bar(df['Country/Territory'][:10], df['2022 Population'][:10]/1000000,)
plt.title('Most Populated Countries [2022]')
plt.xlabel('Countries')
plt.ylabel('Population (in millions)')
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.figure(figsize=(12, 6))

plt.bar(df['Country/Territory'][:10], df['2015 Population'][:10]/1000000,)
plt.title('Most Populated country of 2015')
plt.xlabel('Countries')
plt.ylabel('Population')
plt.xticks(rotation = 'horizontal')
plt.show()
```

```python
df.head(4)
```

```python
asia = df[df['Continent'] == 'Asia']
na = df[df['Continent'] == 'North America']
sa = df[df['Continent'] == 'South America']
```

```python
asia.head(3)
```

```python
continents = ['Asia','South America', 'North America']
population_2022 = np.array([asia['2022 Population'].sum(),
                  sa['2022 Population'].sum(),
                  na['2022 Population'].sum()])

plt.figure(figsize=(12, 6))
plt.bar(continents, population_2022/1000000)
plt.title('Content-wise Population [2022]')
plt.xlabel('Continents')
plt.ylabel('Population (in millions)')
plt.xticks(rotation = 90)
plt.show()
```
